```python
import pandas as pd
import glob

!mkdir biosample_sra

biosample = pd.DataFrame(columns=[
    'replicate','description','*geo_loc_name',
    '*collection_date','sample_title','*organism','*env_medium',
    '*lat_lon','*env_broad_scale','*env_local_scale','treatment','timepoint','original_id','depth'])

biosample.index.name = '*sample_name'


organism = 'water metagenome'
environmental_medium = 'artificial seawater'
broad_scale_environmental_context = 'marine seagrass meadow'

metadata = pd.read_csv('metadata/metadata.tsv',sep='\t',index_col=0)
metadata = metadata.loc[metadata.Fraction=='water']

samples = [i.split('/')[-1].split('_')[0] for i in glob.glob('raw_reads/*.fastq.gz')]

samples = set(samples)

replicate = 'biological replicate %i'

replicate_counter = {}

description = 'Water metagenome from a Halophila stipulacea mesocosm. Treatment %s, time point %s'

collection_dates = ['15-Aug-2019',0,0,'08-Sep-2019',0,'22-Sep-2019',0,0,'13-Oct-2019']

for smpl in samples:
    if smpl not in metadata.index.tolist():
        continue
        
    geo_loc_name = 'Israel:Hatzeva,DeadSeaAndAravaScienceCenter'
    lat_lon = '30.77 N 35.24 E'
    local_scale_environmental_context = "mesocosm experiment"
    collection_date = collection_dates[int(metadata.at[smpl,'TP'])]
    timepoint = metadata.at[smpl,'TP']
    treatment = metadata.at[smpl,'Treatment']
    desc = description % (treatment,timepoint)
    
    
    if desc in replicate_counter:
        replicate_counter[desc] += 1
    else:
        replicate_counter[desc] = 1
        
    rep = replicate%replicate_counter[desc]
    
    
    
    
    biosample.loc[smpl] = [
                    rep,
                    desc +', '+ rep,
                    geo_loc_name,
                    collection_date,
                    desc +', '+ rep,
                    organism,
                    environmental_medium,
                    lat_lon,
                    broad_scale_environmental_context,
                    local_scale_environmental_context,
                    treatment,
                    timepoint,smpl,'9m']
            
for smpl in samples:
    if smpl not in biosample.index:
        print(smpl)
        
biosample.to_csv('biosample_sra/BioSample_metadata_water.tsv',sep='\t')
biosample
```

```python
import glob

sra = pd.DataFrame(columns = [
    'library_ID','title','library_strategy',
    'library_source','library_selection','library_layout',
    'platform','instrument_model','design_description',
    'filetype','filename','filename2']
)

sra.index.name = 'sample_name'

for smpl in biosamples.index:
    biosample = smpl #sample_to_biosample[smpl]
    library_ID = '%s' % smpl
    title = biosamples.loc[biosamples.original_id==smpl]['description'].tolist()[0].replace('metagenome sample','16S-rRNA metabaecoding sequences')+' original ID %s' % biosample
    library_strategy = 'AMPLICON'
    library_source = 'GENOMIC'
    library_selection = 'PCR'
    library_layout = 'paired'
    platform = 'ILLUMINA'
    instrument_model = 'Illumina iSeq 100'
    design_description = 'Single barcode, two-step PCR, V4 fragment'
    filetype = 'fastq'
    R1 = list(glob.glob('raw_reads/%s_*_*_R1_*.fastq.gz'%smpl))[0].split('/')[-1]
    R2 = list(glob.glob('raw_reads/%s_*_*_R2_*.fastq.gz'%smpl))[0].split('/')[-1]
    sra.loc[library_ID] = [biosample,title,library_strategy,library_source,
                           library_selection,library_layout,platform,
                           instrument_model,design_description,filetype,R1,R2]

    
    
sra.to_csv('biosample_sra/sra_metadata_epiphytes.tsv',sep='\t')
```
